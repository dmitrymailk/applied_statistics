```python
import numpy as np
import matplotlib.pyplot as plt
import prettytable as pt
from mpl_toolkits.mplot3d import Axes3D
from IPython.display import set_matplotlib_formats

# specify parameters
success_amount, success_prob, size = 1, 0.0005, 1_000_000

# draw observations from the distribution
x = np.random.negative_binomial(success_amount, success_prob, size)

# compute sample mean and variance
μ_hat = np.mean(x)
σ2_hat = np.var(x)

print("The sample mean is: ", μ_hat, "\nThe sample variance is: ", σ2_hat)
print("\nThe population mean is: ", success_amount*(1-success_prob)/success_prob)
print("The population variance is: ", success_amount*(1-success_prob)/success_prob**2)
```

```python
plt.hist(x, bins=100, density=True)
plt.show()
```
