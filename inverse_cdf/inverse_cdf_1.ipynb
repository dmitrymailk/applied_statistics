```python
import numpy as np
import matplotlib.pyplot as plt
import prettytable as pt
from mpl_toolkits.mplot3d import Axes3D
from IPython.display import set_matplotlib_formats
```

```python
n, λ = 1_000_000, 0.3

# выполняем семпл из Uniform distribution
u = np.random.rand(n)

# трансформируем Uniform distribution в geometric distribution
# путем U = F(x)
# U = 1 - e^(-λX)
# ...
# X = (1 - U) / -λ
uniform_transformed = -np.log(1-u)/λ

# настоящее geometric distributions, 
# которое мы можем взять из готовой функции
true_geometric = np.random.exponential(1 / λ, n)

# plot and compare
plt.hist(uniform_transformed, bins=100, density=True)
plt.show()
```

```python
plt.hist(true_geometric, bins=100, density=True, alpha=0.6)
plt.show()
```

как и ожидалось распределения совпадают
